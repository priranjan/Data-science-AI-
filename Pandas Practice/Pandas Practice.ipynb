# Group By in Pandas

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
#group by continenet and find out mean
drinks.groupby('continent').beer_servings.mean()
```

```python
print(drinks[drinks.continent== 'Africa'].beer_servings.mean())
print(drinks[drinks.continent== 'Africa'].beer_servings.min())
print(drinks[drinks.continent== 'Africa'].beer_servings.max())
print(drinks[drinks.continent== 'Africa'].beer_servings.agg(['count','max','min','mean']))
print(drinks.groupby('continent').beer_servings.agg(['count','max','min','mean']))
```
